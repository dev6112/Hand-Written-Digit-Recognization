```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn import metrics
digits = datasets.load_digits()
X_train, X_test, y_train, y_test = train_test_split(digits.data, digits.target, test_size=1)
clf = svm.SVC(gamma=0.001)
clf.fit(X_train, y_train)
def predict_digit(image):
    flattened_image = image.flatten()
    prediction = clf.predict([flattened_image])
    return prediction[0]
random_index = np.random.randint(0, len(X_test))
random_digit_image = X_test[random_index].reshape(8, 8)
plt.imshow(random_digit_image, cmap=plt.cm.gray_r, interpolation='nearest')
plt.title(f"Actual digit: {y_test[random_index]}")
plt.show()
predicted_digit = predict_digit(X_test[random_index])
print(f"Predicted digit: {predicted_digit}")
```
